### Population Stability Index: Preproceesing #### 

```python
loan_data['emp_length'].unique()
```

```python
loan_data['emp_length'] = loan_data['emp_length'].str.replace('+ years', '', regex=False)
loan_data['emp_length'] = loan_data['emp_length'].str.replace('< 1 year', '0', regex=False)
loan_data['emp_length'] = loan_data['emp_length'].str.replace(' years', '', regex=False)
loan_data['emp_length'] = loan_data['emp_length'].str.replace(' year', '', regex=False)
loan_data['emp_length'] = pd.to_numeric(loan_data['emp_length'], errors='coerce')
loan_data['emp_length'].fillna(0, inplace=True)
loan_data['emp_length'] = loan_data['emp_length'].astype(int)
```
